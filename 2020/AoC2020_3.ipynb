In this example, traversing the map using this slope would cause you to encounter 7 trees.

Starting at the top-left corner of your map and following a slope of right 3 and down 1, how many trees would you encounter?

```python
# Prepare an empty list for our data
data = []

# Open the file as f, pull one line at a time, strip it of the \n at the end of each line, and append it to our data. 
with open("AoC2020_3.txt") as f:
    for line in f.readlines():
        data.append(line.strip())
```

```python
print(len(data))
print(len(data[0]))
```

Our map is 323 entries long and 31 entries wide. 

For the first challenge of a right 3, down 1 slope, we will need 323 steps to reach the bottom. 

```python
j = 0
trees = 0
for i in range(0, len(data)):
    
    j = j % len(data[0])
    
    if data[i][j] == '#':
        trees += 1

    j = j + 7
```
