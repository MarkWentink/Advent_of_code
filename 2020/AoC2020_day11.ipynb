Your plane lands with plenty of time to spare. The final leg of your journey is a ferry that goes directly to the tropical island where you can finally start your vacation. As you reach the waiting area to board the ferry, you realize you're so early, nobody else has even arrived yet!

By modeling the process people use to choose (or abandon) their seat in the waiting area, you're pretty sure you can predict the best place to sit. You make a quick map of the seat layout (your puzzle input).

Now, you just need to model the people who will be arriving shortly. Fortunately, people are entirely predictable and always follow a simple set of rules. All decisions are based on the number of occupied seats adjacent to a given seat (one of the eight positions immediately up, down, left, right, or diagonal from the seat). The following rules are applied to every seat simultaneously:

If a seat is empty (L) and there are no occupied seats adjacent to it, the seat becomes occupied.
If a seat is occupied (#) and four or more seats adjacent to it are also occupied, the seat becomes empty.
Otherwise, the seat's state does not change.
Floor (.) never changes; seats don't move, and nobody sits on the floor.

```python
def count_adjacent(data, i, j):
    occupied = 0
    for x in range(i-1, i+2):
        for y in range(j-1, j+2):
            #print(x, y)
            if x == i and y == j:
                #print("the seat itself")
                pass
            else:
                
                if x >= 0 and x < len(data) and y >= 0 and y < len(data[0]):
                    #print(data[x][y])
                    if data[x][y] == '#':
                        occupied += 1
                        #print("Occupied")
                    else:
                        #print("Floor or vacant")
                        pass
                else:
                    #print("out of bounds")
                    pass

    return occupied
```

```python
changes = 1
while(changes > 0):
    changes = 0
    next_state = copy.deepcopy(seating)
    for i in range(len(seating)):
        for j in range(len(seating[0])):
            if seating[i][j] == 'L' and count_adjacent(seating, i,j) == 0:
                next_state[i][j] = '#'
                changes += 1
            elif seating[i][j] == '#' and count_adjacent(seating, i,j) > 3:
                next_state[i][j] = 'L'
                changes += 1
            else:
                pass
    seating = copy.deepcopy(next_state)
    print(changes)
```

```python
occupied = 0
for row in seating:
    occupied += row.count('#')
```

Given the new visibility method and the rule change for occupied seats becoming empty, once equilibrium is reached, how many seats end up occupied?

```python
from aocd import data
data = data.split('\n')
seating = []
for entry in data:
    row = []
    for seat in entry:
        row.append(seat)
    seating.append(row)
```

```python
def count_visible(data, i, j):
    occupied = 0
    
    for x in range(-1, +2):
        for y in range(-1, +2):
            #print(x, y)
            if x+i == i and y+j == j:
                #print("the seat itself")
                pass
            else:
                if x+i >= 0 and x+i < len(data) and y+j >= 0 and y+j < len(data[0]):
                    n = 1
                    while data[(x*n)+i][(y*n)+j] == '.':
                        if (x*(n+1))+i >= 0 and (x*(n+1))+i < len(data) and (y*(n+1))+j >= 0 and (y*(n+1))+j < len(data[0]):
                            n += 1
                        else:
                            #print("next seat out of bounds")
                            break
                    
                    if data[(x*n)+i][(y*n)+j] == 'L':
                        #print((x*n)+i, (y*n)+j, "empty seat")
                        pass
                    elif data[(x*n)+i][(y*n)+j] == '#':
                        #print((x*n)+i, (y*n)+j, "Occupied")
                        occupied += 1            
                
                else:
                    #print("out of bounds")
                    pass   
    
    return occupied
```

```python
changes = 1
while(changes > 0):
    changes = 0
    next_state = copy.deepcopy(seating)
    for i in range(len(seating)):
        for j in range(len(seating[0])):
            if seating[i][j] == 'L' and count_visible(seating, i,j) == 0:
                next_state[i][j] = '#'
                changes += 1
            elif seating[i][j] == '#' and count_visible(seating, i,j) > 4:
                next_state[i][j] = 'L'
                changes += 1
            else:
                pass
    seating = copy.deepcopy(next_state)
    print(changes)
```

```python
occupied = 0
for row in seating:
    occupied += row.count('#')
```
